```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Find null values
df.isnull().sum()
```

```python
# Drop unnamed column
df = df.drop("Unnamed: 0", axis = 1)
df
```

```python
#change column values
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("5-oct","5-10")
df
```

```python
#Gender distribution

plt.figure(figsize = (5,5))
ax = sns.countplot(data=df , x = "Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#from the above chart we analyze that:
#the numder of female is more than the number of male
```

```python
G1 = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":'mean',"WritingScore":'mean'})
G1
```

```python
sns.heatmap(G1, annot = True)

plt.title("Realtion between Parent education and student scores")
plt.figure(figsize=(4,4))
plt.show()
```

```python
#from above anylasis we notice that:
#parent education effect to student score
```

```python
G2 = df.groupby("PracticeSport").agg({"MathScore":"mean","ReadingScore":'mean',"WritingScore":'mean'})
G2
```

```python
sns.heatmap(G2, annot = True)
plt.show()
plt.figure(figsize=(4,4))
```

```python
# from above analysis notice that:
# Practice sport effect positivily in student score
```

Distibution of Ethnic Groups


```python
df["EthnicGroup"].unique()
```

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

elist = [groupA["EthnicGroup"],groupB["EthnicGroup"], groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
l = ["Group A","Group B","Group C","Group D","Group E"]
plt.pie(elist,labels = l ,autopct = "%1.1f%%")
plt.title("Distribution of Ethnic group")
plt.show()
```
